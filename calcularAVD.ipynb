```python
import numpy as np
import pandas as pd
from datetime import datetime
from tqdm import tqdm_notebook as tqdm
```

```python
pestanasBD = ['General','Accidentes de transporte','Lesiones por agresiones','Autolesiones']
```

```python
pestanaBD = pestanasBD[3]
```

```python
dfPD = pd.read_excel('Libro PD 062021.xlsx', sheet_name='Final', header=2)
dfPD
```

```python
dfBD = pd.read_excel('Libro BD RIPS morbilidad.xlsx', sheet_name=pestanaBD, header=0)
dfBD=dfBD.fillna(0)
dfBD
```

## Calcular AVD por CIE a partir de PD y BD


### Cabecera

```python
dfAVD = dfPD.head(0).drop(columns=['CIE','Description', 'peso'])
dfAVD = pd.concat([dfAVD.head(0), dfBD.head(0)], axis=1)

for i in range(0, 3):
    for j in range(0, 2):
        #print(j)
        dfAVD.at[i, dfAVD.columns[j]] = dfPD.iloc[i][dfAVD.columns[j]]
    for j in range(2, dfBD.shape[1]+2):
    #    print(j)
        dfAVD.at[i, dfAVD.columns[j]] = dfBD.iloc[i][dfAVD.columns[j]]
    
dfAVD
```

### Recorrer todas las filas y columnas en BD a partir de CIE de PD

```python
with tqdm(total=dfPD.shape[0]) as pbar:
    for indexPD, rowPD in dfPD.iterrows():
        if pd.isnull(rowPD['CIE']):
            dfAVD.at[indexPD, dfAVD.columns[0]] = dfPD.iloc[indexPD][dfAVD.columns[0]]
            dfAVD.at[indexPD, dfAVD.columns[1]] = dfPD.iloc[indexPD][dfAVD.columns[1]]
        exists = False
        for indexBD, rowBD in dfBD.iterrows():  
            if rowPD['CIE'] == rowBD['CIE']:
                ncolsBD=len(rowBD)
                dfAVD.at[indexPD, dfAVD.columns[0]] = rowPD[0]
                dfAVD.at[indexPD, dfAVD.columns[1]] = rowPD[1]
                dfAVD.at[indexPD, dfAVD.columns[2]] = rowPD[2]
                dfAVD.at[indexPD, dfAVD.columns[3]] = rowPD[3]
                for i in range(2, ncolsBD):
                    dfAVD.at[indexPD, dfAVD.columns[i+2]] = rowPD['peso']*(rowBD[i])
                exists = True
                break
        if not exists:
            dfAVD.at[indexPD, dfAVD.columns[0]] = dfPD.iloc[indexPD][dfAVD.columns[0]]
            dfAVD.at[indexPD, dfAVD.columns[1]] = dfPD.iloc[indexPD][dfAVD.columns[1]]
            dfAVD.at[indexPD, dfAVD.columns[2]] = dfPD.iloc[indexPD][dfAVD.columns[2]]
            dfAVD.at[indexPD, dfAVD.columns[3]] = dfPD.iloc[indexPD][dfAVD.columns[3]]
        pbar.update()
        pbar.set_description("Procesando " + str(rowPD['CIE']))
    pbar.close()
```

## Guardar en archivo Excel

```python
now = datetime.now() # current date and time
date_time = now.strftime("%Y-%m-%d")#_%H-%M-%S")
print("date and time:",date_time)
dfAVD.to_excel("Libro AVD 062021-"+date_time+".xlsx", sheet_name=pestanaBD)  
```
